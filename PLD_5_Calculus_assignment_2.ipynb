# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
#TO DO: Create a function that does a derivative for any function
def calculate_derivative(func_str, var='x'):
    x = sp.symbols(var)
    func = sp.sympify(func_str)
    derivative = sp.diff(func, x)
    return derivative
````

# Task 2: Test the derivative function written with an equation of your choice
*NB: Must have atleast 2 minimas and atleast 2 maximas*

````python
from sympy import symbols, diff, sin

x = symbols('x')
func_str = "x**4 - 6*x**2 + 8"
func = sp.sympify(func_str)
derivative_func = calculate_derivative(func_str)
print("Derivative:", derivative_func)
````

#Task 3: Plot a graph of the Equation must have 4 points (2 minimas and 2 maximas)

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x_vals = np.linspace(-3, 3, 400)
y_vals = [func.subs(x, val) for val in x_vals]  # Compute corresponding y values

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x)")
plt.title("Graph of the Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
````

# Task 4: Create separate array of Minimas and maximas

````python
critical_points = sp.solve(derivative_func, x)
critical_points = [point.evalf() for point in critical_points]

minima = []
maxima = []

# Determine concavity at critical points
second_derivative = sp.diff(derivative_func, x)
for point in critical_points:
    concavity = second_derivative.subs(x, point).evalf()
    if concavity > 0:
        minima.append((point, func.subs(x, point)))
    elif concavity < 0:
        maxima.append((point, func.subs(x, point)))

print("Minima:", minima)
print("Maxima:", maxima)
````

````python
if minima:
    global_min = min(minima, key=lambda p: p[1])
else:
    global_min = None

if maxima:
    global_max = max(maxima, key=lambda p: p[1])
else:
    global_max = None

print("Global Minima:", global_min)
print("Global Maxima:", global_max)
````

#Finally, What is the global Minima and the Global maxima _ Plot this so that I can see

````python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x)")

# Mark global minima and maxima
if global_min:
    plt.scatter(*zip(*[global_min]), color='g', marker='o', label=f"Global Min ({global_min[0]:.2f}, {global_min[1]:.2f})")
if global_max:
    plt.scatter(*zip(*[global_max]), color='r', marker='x', label=f"Global Max ({global_max[0]:.2f}, {global_max[1]:.2f})")

plt.title("Function with Global Minima and Maxima")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
````
